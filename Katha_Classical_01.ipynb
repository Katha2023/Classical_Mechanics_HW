```python
from matplotlib import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
k = 500;
x1=100;
T = np.arange(0,4.8,0.01); X = [];
for t in T:
    x= x1-0.5*9.8*(t**2);
    X.append(x)  
plt.plot(T,X)
plt.ylim(0,110)
plt.xlim(0,5)
plt.xlabel("time(s)")
plt.ylabel("Height(m)")
plt.title('Falling Object')
```
