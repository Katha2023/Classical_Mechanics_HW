# Solving with Eular 

```python
from __future__ import division
import numpy as np
import scipy.integrate
from numpy import *
import matplotlib.pyplot as plt
```

```python
m = 0.1
k = 0.4
x0 = 0.5
v0 = 0.0
t0 = 0.0
```

```python
t1 = 35
n_steps = 500
dt = (t1-t0)/n_steps
```

```python
xs = np.zeros((n_steps),)
vs = np.zeros((n_steps),)
ts = np.zeros((n_steps),)
```

```python
def dx_dt(x,v):
    return v
def dv_dt(x,v):
    return -(k/m)*x

```

```python
for i in range(n_steps-1):
    k0x = dx_dt(x0,v0)
    k0v = dv_dt(x0,v0)
    x1 = x0+k0x*dt
    v1 = v0+k0v*dt
    vs[i+1] = v1
    xs[i+1] = x1
    ts[i+1] = t0+i*dt
    v0 = v1
    x0 = x1
```

```python
plt.figure(figsize=(8,8))
plt.subplot(211)
plt.plot(ts, xs, color="red")
plt.xlabel('time(s)')
plt.ylabel('position(m)')
plt.subplot(212)
plt.plot(xs, vs, color="red")
plt.xlabel('position(m)')
plt.ylabel('velocity(m/s)')

plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(211)
plt.plot(ts, xs, color="blue")
plt.xlabel('time(s)')
plt.ylabel('position(m)')
plt.subplot(212)
plt.plot(xs, vs, color="blue")
plt.xlabel('position(m)')
plt.ylabel('velocity(m/s)')

plt.show()
```
